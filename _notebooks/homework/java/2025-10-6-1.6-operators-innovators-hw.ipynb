### **Option B: Bank Account Manager**
Simulate managing a personal bank account over time:

##### **Suggested variables to track:**
- `balance` - Current account balance
- `transactions` - Number of transactions made
- `monthlyFee` - Account maintenance costs
- `interestRate` - Interest earned/applied

##### **Example operations you might use:**
- Add monthly salary (`balance += 3000`)
- Subtract rent payment (`balance -= 1200`)
- Apply interest rate (`balance *= 1.02`)
- Average monthly spending (`spending /= 12`)
- Check account tier eligibility (`tier %= 5`)

```java
class BankAccount {
    private double checkingBalance;
    private double savingsBalance;
    private String accountNumber;

    public BankAccount() {
        this.checkingBalance = 0.0;
        this.savingsBalance = 0.0;
        this.accountNumber = "123456789";
    }

    public void deposit(double amount, String accountType) {
        if (accountType.equals("checking")) {
            System.out.println("Depositing $" + amount + " to checking account.");
            this.checkingBalance += amount;
        } else if (accountType.equals("savings")) {
            System.out.println("Depositing $" + amount + " to savings account.");
            this.savingsBalance += amount;
        } else {
            System.out.println("Invalid account type.");
        }
    }

    public double withdraw(double amount, String accountType) {
        if (accountType.equals("checking")) {
            if (amount <= this.checkingBalance) {
                System.out.println("Withdrawing $" + amount + " from checking account.");
                this.checkingBalance -= amount;
                return amount;
            } else {
                System.out.println("Insufficient funds in checking account.");
                return 0.0;
            }
        } else if (accountType.equals("savings")) {
            if (amount <= this.savingsBalance) {
                System.out.println("Withdrawing $" + amount + " from savings account.");
                this.savingsBalance -= amount;
                return amount;
            } else {
                System.out.println("Insufficient funds in savings account.");
                return 0.0;
            }
        } else {
            System.out.println("Invalid account type.");
            return 0.0;
        }
    }

    public void interestRate() {
        System.out.println("Applying 1% interest to savings account.");
        this.savingsBalance *= 1.01;
    }

    public double getCheckingBalance() {
        System.out.println("Checking Balance: $" + this.checkingBalance);
        return this.checkingBalance;
    }

    public double getSavingsBalance() {
        System.out.println("Savings Balance: $" + this.savingsBalance);
        return this.savingsBalance;
    }
}

BankAccount myAccount = new BankAccount();
myAccount.deposit(1000.0, "checking");
myAccount.deposit(5000.0, "savings");
myAccount.getCheckingBalance();  
myAccount.withdraw(200.0, "checking");
myAccount.getCheckingBalance();
myAccount.getSavingsBalance();  
myAccount.interestRate();
myAccount.getSavingsBalance();  
```

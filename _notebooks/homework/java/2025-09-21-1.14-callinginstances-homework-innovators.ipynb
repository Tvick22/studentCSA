````java
import java.text.DecimalFormat;
import java.util.ArrayList;

public class Student {
    // Required instance variables
    private String name;
    private int totalPoints;
    private int numAssignments;

    // Optional: store individual grades for bonus features (not required)
    private ArrayList<Integer> grades;

    // Constructor
    public Student(String name) {
        this.name = name;
        this.totalPoints = 0;
        this.numAssignments = 0;
        this.grades = new ArrayList<>();
        System.out.println("Creating student: " + name);
        System.out.println("Student created successfully!\n");
    }

    // Void instance method that modifies the student's grades
    public void addGrade(int points) {
        if (points < 0) {
            System.out.println("Invalid grade (negative). Grade ignored: " + points);
            return;
        }
        System.out.println("Adding grade: " + points + " points");
        this.totalPoints += points;
        this.numAssignments++;
        this.grades.add(points);
    }

    // Non-void instance method: returns average as a double
    public double getAverage() {
        if (this.numAssignments == 0) return 0.0;
        return (double) this.totalPoints / this.numAssignments;
    }

    // Non-void instance method: returns letter grade
    public String getLetterGrade() {
        double avg = getAverage();
        if (this.numAssignments == 0) return "N/A";
        if (avg >= 90) return "A";
        if (avg >= 80) return "B";
        if (avg >= 70) return "C";
        if (avg >= 60) return "D";
        return "F";
    }

    // Void instance method: prints a formatted report for this student
    public void printReport() {
        DecimalFormat df = new DecimalFormat("#.##"); // formats like 78.33 or 90
        System.out.println("\n--- " + this.name + "'s Grade Report ---");
        System.out.println("Student Name: " + this.name);
        System.out.println("Total Points: " + this.totalPoints);
        System.out.println("Assignments Completed: " + this.numAssignments);
        if (this.numAssignments == 0) {
            System.out.println("Current Average: N/A");
            System.out.println("Letter Grade: N/A");
            System.out.println("Status: No grades recorded yet.");
        } else {
            double avg = getAverage();
            System.out.println("Current Average: " + df.format(avg));
            System.out.println("Letter Grade: " + getLetterGrade());
            System.out.println("Status: " + statusMessage(avg));
        }
        System.out.println("\n========================================\n");
    }

    // Helper to produce a small status message based on average
    private String statusMessage(double avg) {
        if (avg >= 90) return "Excellent work!";
        if (avg >= 80) return "Great job — keep it up!";
        if (avg >= 70) return "Keep working hard!";
        if (avg >= 60) return "Room for improvement.";
        return "Needs significant improvement. Ask for help!";
    }

    // Bonus (optional): getHighestGrade
    public int getHighestGrade() {
        if (grades.isEmpty()) return -1;
        int max = grades.get(0);
        for (int g : grades) if (g > max) max = g;
        return max;
    }

    // Bonus (optional): getLowestGrade
    public int getLowestGrade() {
        if (grades.isEmpty()) return -1;
        int min = grades.get(0);
        for (int g : grades) if (g < min) min = g;
        return min;
    }

    // Main method to demonstrate at least 2 Student instances and all methods
    public static void main(String[] args) {
        System.out.println("=== Student Grade Tracker System ===\n");

        Student trevor = new Student("Trevor Vick");
        System.out.println("--- Adding Grades for Trevor ---");
        trevor.addGrade(95);
        trevor.addGrade(88);
        trevor.addGrade(92);
        trevor.addGrade(85);
        trevor.printReport();

        Student john = new Student("John Doe");
        System.out.println("--- Adding Grades for John ---");
        john.addGrade(78);
        john.addGrade(82);
        john.addGrade(75);
        john.printReport();

        // Final Summary
        DecimalFormat df = new DecimalFormat("#.##");
        System.out.println("Final Summary:");
        System.out.println(trevor.name + " - Average: " + (trevor.numAssignments == 0 ? "N/A" : df.format(trevor.getAverage()))
                           + " (" + trevor.getLetterGrade() + ")");
        System.out.println(john.name + " - Average: " + (john.numAssignments == 0 ? "N/A" : df.format(john.getAverage()))
                           + " (" + john.getLetterGrade() + ")");

        // Demonstrate bonus methods (optional)
        System.out.println("\n(Extras) Highest and Lowest recorded grades:");
        System.out.println(trevor.name + " - Highest: " + trevor.getHighestGrade() + ", Lowest: " + trevor.getLowestGrade());
        System.out.println(john.name + " - Highest: " + john.getHighestGrade() + ", Lowest: " + john.getLowestGrade());
    }
}

Student.main(null);
````
